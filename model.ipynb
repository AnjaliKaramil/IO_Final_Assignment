```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import glob
```

```python
filenames=glob.glob(".\datasets\*.csv")
```

```python
filenames
```

Applying the FORWARD FILL method. It is generally apllied to TimeSeries data. NaN values are replaced with the preceding row's value for that column. This can be done because the weather conditions for today can be assumed to be fairly similar to that of yesterday in each city.  

```python
dfs=[] #List to store dataframes read from all files
for file in filenames:
    each=pd.read_csv(file, header=0)
    #Setting Date column as the index of the dataframe
    each.Date = pd.to_datetime(each.Date, format='%Y-%m-%d')
    each.set_index('Date', inplace=True)
    print("Initial Number of NaNs - City = {}".format(each['Location'][0]))
    print(each.isnull().sum())
    each.fillna(method="ffill", inplace=True)
    print("--------------------------")
    print("After Filling, Number of NaNs - City = {}".format(each['Location'][0]))
    print(each.isnull().sum())
    print("**************************")
    dfs.append(each)
```

```python
df=pd.concat(dfs, axis=0 )
#Concatenating all dataframes into one variable.
```

```python
df.info()
```

Note that NaN values still exist in the dataset. <br>
I have chosen to fill the remaining NaNs in all categorical columns with their mode and for all non-categorical, continuous columns with their mean.

```python
df.columns
```

```python
categorical_cols= ['Location', 'WindGustDir', 'WindDir9am', 'WindDir3pm', 'RainToday', 'RainTomorrow']
non_categorical_cols= ['MinTemp', 'MaxTemp', 'Rainfall', 'Evaporation', 'Sunshine', 
        'WindGustSpeed', 'WindSpeed9am', 'WindSpeed3pm', 'Humidity9am', 'Humidity3pm',
       'Pressure9am', 'Pressure3pm', 'Cloud9am', 'Cloud3pm', 'Temp9am', 'Temp3pm',
        'RISK_MM']
```

```python
#Replacing the NULLS in categorical columns with their modes:
#For example:
print(df['WindGustDir'].mode())
#Pandas Series object returned: First row holds the index & the mode 
```

```python
for col in categorical_cols:
    df[col].fillna(df[col].mode()[0], inplace=True)
```

```python
#All the categorical columns now have no missing values:
df.isnull().sum()
```

```python
#Replacing the NULLS in non-categorical float columns with their means:
#For example:
print(df['MinTemp'].mean())
#Float returned 
```

```python
for col in non_categorical_cols:
    df[col].fillna(df[col].mean(), inplace=True)
```

```python
#There are no null values in df now:
df.isnull().sum()
```

Location, WindGustDir, WindDir9am, WindDir3pm, RainToday, RainTomorrow are columns containing string categorical data. These have to be indexed to integers as the ML models can process only integer data.

```python
#Converting the column types to 'Category' types as they contain categorical data:
for col in categorical_cols:
    df[col]=df[col].astype('category')
```

```python
df[categorical_cols] = df[categorical_cols].apply(lambda x: x.cat.codes)
#All categorical columns have an attribute df.col.cat.code which
#gives a series same as df.col except every category  name in df.col is replaced 
#with a unique integer from 0, 1, ... no_of_categories-1.
```

Right now, all the columns are either float or int type. It is now ready for our model.

I have chosen the ANN model

### ANN Model:

```python
import tensorflow as tf
#TensorFlow's eager execution is a programming environment that evaluates operations immediately, without building graphs
#This makes it easier to debug models.
```

```python
from sklearn.model_selection import train_test_split
```

```python
train, test = train_test_split(df, test_size=0.2)
print(len(train), 'train examples')
print(len(test), 'test examples')
```

```python
#Making Tensorflow datasets corresponding to train and test pandas dataframes:
copy=train.copy() #Making a deep copy of train dataframe
labels = copy.pop('RainTomorrow')  #Rain Tomorrow is the label column
train_ds = tf.data.Dataset.from_tensor_slices((dict(train), labels))
```

```python
#Similarly for test dataframe:
copy=test.copy() 
labels = copy.pop('RainTomorrow')  
test_ds = tf.data.Dataset.from_tensor_slices((dict(test), labels))
```

```python
#Shuffling the train and test datasets:
train_ds = train_ds.shuffle(buffer_size=len(train))
test_ds =  test_ds.shuffle(buffer_size=len(test))
#Setting the buffer_size equal to the number of records. 
#The buffer_size refers to the pool size from which records will be randomly selected.
```

```python
#Dividing the rows in the datasets into batches:
train_ds = train_ds.batch(32) #batch size chosen as 32 usually.
test_ds = test_ds.batch(32)
#The batch size determines how many rows will be processed at one time.
```

```python
from tensorflow import feature_column
#A feature_column is created which maps columns in the dataframe to features used to
#train the model
```

```python
feature_columns = []
features=list(df.columns)[:-1]  #features is a list of all column names other than RainTomorrow
label=list(df.columns)[-1]    #RainTomorrow is the target column
```

```python
features
```

```python
label
```

```python
for col in features:
    feature_columns.append(feature_column.numeric_column(col))  
#All the columns have been converted to numeric type data. Hence, numeric_column() method can be used.
```

```python
from tensorflow.keras import layers
```

```python
feature_layer = tf.keras.layers.DenseFeatures(feature_columns)
#Produces a dense tensor based on the given feature columns
```

```python
model = tf.keras.Sequential([
  feature_layer,
  layers.Dense(128, activation='relu'),
  layers.Dense(128, activation='relu'),
  layers.Dense(1, activation='sigmoid')
])
#The ANN has an input layer containing number of neurons equal to the number of features.
#Two hidden layers having 128 neurons each. ReLU is used as the activation function.
#The output layer has only one neuron because there are only two possible outputs - either
#'It will rain tomorrow' or 'It will not rain tomorrow'.
```

```python
model.compile(optimizer='adam',
              loss='binary_crossentropy',
              metrics=['accuracy'],
              run_eagerly=True)
#Adam optimizer is used.
#Loss function used is binary_crossentropy function.
#Evaluating only accuracy metrics
#Setting run_eagerly=True so that model becomes easier to debug
```

```python
model.fit(train_ds,
          epochs=1)
```

```python
loss, accuracy = model.evaluate(test_ds)
print("Accuracy", accuracy)
```
